```python
import matplotlib.pyplot as plt

years = [2016, 2026]
demand = [650000, 2500000]  # in tonnes

plt.figure(figsize=(8,6))
plt.plot(years, demand, marker='o', linestyle='-', color='blue')
plt.title('Projected Copper Demand in India (2016-2026)')
plt.xlabel('Year')
plt.ylabel('Copper Demand (in tonnes)')
plt.grid(True)
plt.show()
```
