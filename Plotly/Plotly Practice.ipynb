# TASK #1: PLOT INTERACTIVE SCATTERPLOT USING PLOTLY EXPRESS

```python
import plotly.express as px
import pandas as pd

salary_df = px.df = pd.read_csv('/workspaces/Coding/Plotly/employee_salaries.csv')
display(salary_df)
# Plot Years of Experience Vs. Salary Using Plolty Express
chart = px.scatter(salary_df, x = 'Years_of_Experience', y = 'Salary', color_discrete_sequence = px.colors.qualitative.Alphabet)
chart.show()
```

# TASK #2: PLOT INTERACTIVE BUBBLE CHART (SCATTERPLOT WITH SIZE)

```python
# Let's add a fourth variable "SOP" as the size 
chart = px.scatter(admission_df, x = "GRE Score", y = "Chance of Admit", color = "University Rating", size = "SOP")
chart.show()
```

```python
chart = px.scatter(admission_df, x = "GRE Score", y = "Chance of Admit", color = "University Rating", size = "SOP", hover_data = 'LOR')
chart.show()
```
